```python
# Salary distribution
plt.figure(figsize=(6,6))
sns.histplot(data['Salary'])
plt.show()
```

```python
# Salary Distribution
plt.figure(figsize=(6,6))
sns.distplot(data['Salary'])
plt.show()
```

```python
# Employee Satisfaction

plt.hist(data['EmpSatisfaction'])
plt.xlabel("Job Satisfaction")
plt.ylabel("count")
plt.show()
```

```python
# Different Departments

plt.subplots(figsize=(10,5))
plt.hist(data['Department'])
plt.xlabel("Department")
plt.ylabel("count")
plt.show()
```

```python
# Bivariate Analysis

plt.scatter(data['Salary'], data['Department'])
plt.title('Salary Vs Department')
plt.show()
```

```python
# Bivariate Analysis

plt.scatter(data['Salary'], data['PerformanceScore'])
plt.title('Salary Vs Performance Score')
plt.show()
```

```python
# Multivariate Analysis
```

```python
# Correlation Matrix

corrmatrix=data.corr()
plt.subplots(figsize=(20,8))
sns.heatmap(corrmatrix,vmin=-0.4,vmax=0.9,linewidth=0.2,annot=True,cmap='Greens')
```
